Question 9) 
Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample 
size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to 
reject the null hypothesis?

 Task: Write Python code to calculate the z-score and p-value for the given data.

Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing

```python
# Answer 9)

import scipy.stats as stats #zcritical approach
import numpy as np
# as per question :
mu=105
x=100
sigma=15
n=25 
alpha=0.05
# where h0=105 and Ha!=100
zscore=(mu-x)/(x/np.sqrt(50))
print ('the zscore is',zscore)
```

```python
#for zcritical scipy.stats
zcritical = stats.norm.ppf(1-alpha)
zcritical
```

```python
if zscore > zcritical:
    print("Reject the null hypothesis")
else:
    print("fail to reject the null hypothesis")
```

```python
#using bscipy.stats pvalue approach
p_value = 1-stats.norm.cdf(zscore)

if p_value <= 0.05:
    print("reject the null hypothesis")
else:
    print("fail to reject null hypothesis")
```
